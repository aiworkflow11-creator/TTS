```python
import asyncio
import edge_tts
from pathlib import Path

# Read Hindi story from file (uses input/hindi_story.txt)
story_path = Path("/workspaces/TTS/input/hindi_story.txt")
if not story_path.exists():
    raise FileNotFoundError(f"{story_path} not found. Place the file at input/hindi_story.txt")

TEXT = story_path.read_text(encoding="utf-8")
TEXT = TEXT.replace("[NAME]", "Saurabh")  # Replace newlines with spaces for better TTS results
print(TEXT)
VOICE = "hi-IN-SwaraNeural"  # Use MadhurNeural for a male voice
OUTPUT_FILE = "/workspaces/TTS/output/story_audio.mp3"

async def generate_audio():
    communicate = edge_tts.Communicate(TEXT, VOICE)
    await communicate.save(OUTPUT_FILE)
    print(f"Success! Audio saved to {OUTPUT_FILE}")

def run_coro(coro):
    try:
        loop = asyncio.get_running_loop()
    except RuntimeError:
        # No running loop: safe to use asyncio.run() (script mode)
        return asyncio.run(coro)
    else:
        # A loop is already running (e.g. in Jupyter).
        # Schedule the coroutine and return the Task so callers can await it.
        return loop.create_task(coro)

# When running as a script this will execute; in notebooks prefer `await generate_audio()`
if __name__ == "__main__":
    run_coro(generate_audio())
```
